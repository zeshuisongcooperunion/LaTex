```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
# Data for Youngs (lengths in nanometers)
d = 1000 #1nm
lamb = 650.0
# Data for plotting
#theta1 = np.arange(-np.pi/2.0, np.pi/2.0, 0.001)#
theta = np.linspace(-np.pi/2.0, np.pi/2.0, int(np.pi/0.001))
delta = (2.0 * np.pi / lamb) * d * np.sin(theta)
bigI = (np.cos(delta/2.0))**2
dovlamsintheta = (d / lamb) * np.sin(theta)
fig, ax = plt.subplots()
ax.plot(dovlamsintheta, bigI, color='red')
ax.set(xlabel='position (d/lambda)*sin(theta)', ylabel='Intensity (I)',
title='Youngs Double Slit Intensity')
ax.grid()
fig.savefig("youngs_ds.png")
plt.show()
```

```python
deg = np.rad2deg(theta)
delta_n = (2.0 * np.pi / lamb) * d * np.sin(theta)

N = 4
I0 = 1/N**2
I4 = I0 * (np.sin(N*delta_n/2.0)/np.sin(delta_n/2.0))**2
I4 = np.where(np.isfinite(I4), I4, 1.0) 

N = 16
I0 = 1/N**2
I16 = I0 * (np.sin(N*delta_n/2.0)/np.sin(delta_n/2.0))**2
I16 = np.where(np.isfinite(I16), I16, 1.0)

N = 256
I0 = 1/N**2
I256 = I0 * (np.sin(N*delta_n/2.0)/np.sin(delta_n/2.0))**2
I256 = np.where(np.isfinite(I256), I256, 1.0)

fig, ax = plt.subplots()
plt.xlim(-25, 25) 
ax.plot(deg, I4, color='red', label='N=4')
ax.plot(deg, I16, color='blue', label='N=16')
ax.plot(deg, I256, color='black', label='N=256')

ax.set(xlabel='$\\theta$ (Degrees) ', ylabel='Normalized Intensity ($I/I_o$)',
title='N-Slit Intensity')
ax.grid()
ax.legend()
fig.savefig("N-Slit.png")
plt.show()
```
